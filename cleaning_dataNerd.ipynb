```python
def number_unique(df, columns):
    '''
    This function returns each column and the number of unique records in it.
    '''
    data = []  # List to store data for each column
    
    for col in columns:
        col_unique = df[col].nunique()
        data.append([col, col_unique])
    
    dataframenunique = pd.DataFrame(data, columns=['feature from data_nerd', 'number of unique record'])
    
    return dataframenunique
```

```python
#defining Position level
def categorize_title(title_original):
    title_lower =title_original.lower()
    title_levels = [
        ('staff', ['director','expert', 'supervisor','lead' 'dir', 'iv','manager','mgr', 'advisor' ,'advanced',  'principal', 'iii']),
        ('senior',['master', 'phd' ,'senior', 'mentor','sr.','mid']),
        ('junior', ['junior', 'jr', '2', 'ii', 'entry']),
        ('intern', ['apprenticeship', 'intern', 'internship',  '1'])]
    
    for title_defined, keywords in title_levels:
        if any(keyword in title_lower for keyword in keywords):
            return title_defined
    return 'any'
```

```python
df['position_level']=df['title'].apply(categorize_title)
```

```python
def categorize_title_types(title):
    # Convert the title name to lowercase for case-insensitive matching
    title_lower = title.lower()


    # Define a list of title levels
    title_types = [
        ('Engineer', ['engineer']),
        ('Architect', ['architect']),
        ('Scientist', ['scientist']),
        ('Analyst', ['analyst']),
        ('Developer', ['developer']),
        ('Programmer', ['programmer']),
        ('Specialist', ['specialist']),
        ('Writer', ['writer']),
    ]

    # Iterate through the conditions and return the industry if a match is found
    for title, keywords in title_types:
          if any(keyword in title_lower for keyword in keywords):
            return title

    # Default to 'Other' if no match is found
    return 'Other'

df['title_types'] = df['title'].apply(categorize_title_types)
```

```python
df.rename(columns={'title_types': 'job_role'}, inplace=True)
```

```python
def categorize_industry(company_name):
    # Convert the company name to lowercase for case-insensitive matching
    company_name_lower = company_name.lower()


    # Define a list of industry conditions
    industry_conditions = [
        ('Technology', ['tech','technology', 'software', 'google', 'meta', 'amazon', 'facebook', 'programmers', 'technologies', 'cisco meraki', 'ust','zoom']),
        ('Finance', ['finance', 'bank', 'credit', 'fpa', 'financial', 'massmutual', 'citigroup']),
        ('Healthcare', ['health', 'medical', 'med', 'hospital', 'memorial','centene']),
        ('Insurance', ['insurance', 'lockton']),
        ('Education', ['college', 'education' 'university']),
        ('Government', ['county', 'city', 'state', 'federal']),
        ('Retail', ['walmart', 'sam','iseatz', 'airlines', 'cbre','hotwire', 'home depot', 'cox']),
        ('Staffing', ['recruit', 'upwork', 'agency', 'staffing', 'talentify.io', 'saxon global', 'flexjobs', 'robert half', 'jobot', 'dice']),
        ('Utilities', ['electric', 'nfuenergy', 'comcast']),
    ]

    # Iterate through the conditions and return the industry if a match is found
    for industry, keywords in industry_conditions:
          if any(keyword in company_name_lower for keyword in keywords):
            return industry

    # Default to 'Other' if no match is found
    return 'Other'

# Apply the function to create a new 'industry' column
df['industry'] = df['company_name'].apply(categorize_industry)
```

```python
def extract_state(location):
    if pd.notnull(location):
        # Splitting by comma and taking the last part (usually the state)
        parts = location.split(',')
        if len(parts) > 1:
            return parts[-1].strip()
    return None

df['state'] = df['location'].apply(extract_state)
df['state'] = df['state'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
# Mapping to reduce categories
category_mapping = {
    'Contractor': 'Contractor',
    'Full-time': 'Full-time',
    'Contractor and Temp work': 'Contractor or Temp Work',
    'Full-time and Part-time': 'Full time',
    'Part-time': 'Part-time',
    'Internship': 'Internship',
    'Temp work': 'Contractor',
    'Full-time Contractor and Temp work': 'Full-time',
    'Full-time and Temp work': 'Full-time',
    'Full-time and Contractor': 'Full-time',
    'Part-time and Internship': 'Full-time',
    'Full-time and Internship': 'Full-time',
    'Full-time Part-time and Internship': 'Full-time',
    'Volunteer': 'Other',
    'Per diem': 'Other',
    'Part-time Contractor and Temp work': 'Part-time',
    'Part-time and Contractor': 'Part-time',
    'Part-time and Temp work': 'Part-time',
    'Full-time Part-time and Temp work': 'Full-time',
    'Full-time and Per diem': 'Full-time',
    'Full-time Part-time and Contractor': 'Full-time',
    'Temp work and Internship': 'Internship',
}

df['contract_type']=df['contract_type'].map(category_mapping)
```
